```python
## importing libraries

import pandas as pd
import numpy as np
import plotly.express as px
```

```python
## loading in dataset

spending=pd.read_csv('https://raw.githubusercontent.com/mawonuga121/randomdata/main/student_spending%20.csv')
spending
```

```python
## data prep I: data cleaning part one (getting rid of import errors)

spending.columns

spending_clean= spending.drop('Unnamed: 0', axis=1)

spending_clean
```

```python
## data prep II: learning about the dataset

spending_clean.info()

#spending_clean.describe()
```

```python
## data prep part III: data cleaning part two (narrowing down our dataset)

spending_final=spending_clean.drop(['preferred_payment_method','entertainment', 'miscellaneous', 'housing', 'food', 'personal_care'], axis=1)
spending_final= spending_final[(spending_final['monthly_income'] <= 950)]
spending_final= spending_final[(spending_final['age'] <= 22)]

spending_final
```

```python
## data analysis part I: statistics

spending_final.describe()
```

```python
## data analysis part II: data visualization featuring plotly

fig1 = px.histogram(spending_final, x="monthly_income", nbins=25)
fig1.update_layout(bargap=0.01)
fig1.show()
```

```python
fig2=px.pie(spending_final, values='financial_aid', names='year_in_school')
fig2.show()
```

```python
fig3 = px.histogram(spending_final, x="major", y="tuition", color="year_in_school", barmode='group', labels={"major": "Major","tuition": "Tuition"})
fig3.show()
```

```python
fig4 = px.scatter(spending_final, x="tuition", y="health_wellness", trendline="ols")
fig4.show()
```
